```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Search for a file among possible extensions
candidates = ['einstein.tiff', 'einstein.tif']
file_path = None
for p in candidates:
    if os.path.exists(p):
        file_path = p
        break
if file_path is None:
    raise FileNotFoundError(f"Изображение Einstein не найдено по путям: {candidates}")
```

```python
# Loading grayscale image
img_gray = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Display original image
plt.figure(figsize=(6,6))
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.title('Оригинал в градациях серого')
plt.show()
```

```python
# Construction of the negative
img_neg = 255 - img_gray
```

```python
# Displaying a negative image
plt.figure(figsize=(6,6))
plt.imshow(img_neg, cmap='gray')
plt.axis('off')
plt.title('Негатив изображения')
plt.show()
```

```python
# Select 5 random pixels and output their intensities
h, w = img_gray.shape
np.random.seed(0)  # для воспроизводимости
ys = np.random.randint(0, h, size=5)
xs = np.random.randint(0, w, size=5)
```

```python
print("Интенсивности пикселей в оригинале и в негативе:")
for y, x in zip(ys, xs):
    orig_val = int(img_gray[y, x])
    neg_val = int(img_neg[y, x])
    print(f'Пиксель (y={y}, x={x}): оригинал = {orig_val}, негатив = {neg_val}')
```
